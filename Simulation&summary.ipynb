```python
def plot_path_with_arrow(points, likelihoods, path):
    # Plot the points and their likelihoods
    labeled_points = set()
    for i, point in enumerate(points):
        plt.scatter(*point, color='red')
        if i not in labeled_points:
            #plt.annotate(f"Point {i}: ({point[0]}, {point[1]})\nLikelihood: {likelihoods[i]:.2f}", xy=point, ha='right', va='bottom', color='black')
            plt.annotate(f"{likelihoods[i]:.2f}", xy=point, ha='right', va='bottom', color='black')
            labeled_points.add(i)

    # Plot the lines with arrowheads
    for i in range(len(path)-1):
        start = points[path[i]]
        end = points[path[i+1]]
        mid = ((start[0]+end[0])/2, (start[1]+end[1])/2)
        dx = end[0] - start[0]
        dy = end[1] - start[1]
        plt.annotate("", xy=start, xytext=end, arrowprops=dict(arrowstyle='-', color='blue'), ha='center')
        plt.arrow(start[0], start[1], dx/2, dy/2, color='blue', head_width=0.1)

    # Set plot limits and labels
    plt.xlim(min(x for x, _ in points) - 1, max(x for x, _ in points) + 1)
    plt.ylim(min(y for _, y in points) - 1, max(y for _, y in points) + 1)
    plt.xlabel('X')
    plt.ylabel('Y')

    # Display the plot
    plt.show()
```

```python
def plot_path_with_arrow_and_distance(points, likelihoods, path):
    # Plot the points and their likelihoods
    labeled_points = set()
    for i, point in enumerate(points):
        plt.scatter(*point, color='red')
        if i not in labeled_points:
            plt.annotate(f"Point {i}: ({point[0]}, {point[1]})\nLikelihood: {likelihoods[i]:.2f}", xy=point, ha='right', va='bottom', color='black')
            labeled_points.add(i)

    # Plot the lines with arrowheads and print distances
    total_distance = 0
    for i in range(len(path)-1):
        start = points[path[i]]
        end = points[path[i+1]]
        mid = ((start[0]+end[0])/2, (start[1]+end[1])/2)
        dx = end[0] - start[0]
        dy = end[1] - start[1]
        distance = np.linalg.norm(end - start)
        total_distance += distance
        plt.annotate(f"  {distance:.2f}", xy=(mid[0]+0.4, mid[1]), ha='center', va='center', color='black')  # Adjust the x-coordinate by adding 0.2
        plt.annotate("", xy=start, xytext=end, arrowprops=dict(arrowstyle='-', color='blue'), ha='center')
        plt.arrow(start[0], start[1], dx/2, dy/2, color='blue', head_width=0.1)

    # Set plot limits and labels
    plt.xlim(min(x for x, _ in points) - 1, max(x for x, _ in points) + 1)
    plt.ylim(min(y for _, y in points) - 1, max(y for _, y in points) + 1)
    plt.xlabel('X')
    plt.ylabel('Y')

    # Display the plot and print total distance
    plt.show()
    print("Total Distance:", total_distance)
```

```python
# Function to calculate the Euclidean distance between two points
import numpy as np
def calculate_distance(point1, point2):
    #return np.linalg.norm(point2 - point1)
    #return math.sqrt((point2[0] - point1[0])**2 + (point2[1] - point1[1])**2)
    # Convert RA and Dec from degrees to radians
    ra1,dec1 = point1[0], point1[1]
    ra2, dec2 = point2[0], point2[1]
    ra1 = np.radians(ra1)
    dec1 = np.radians(dec1)
    ra2 = np.radians(ra2)
    dec2 = np.radians(dec2)

    # Haversine formula
    dlon = ra2 - ra1
    dlat = dec2 - dec1
    a = np.sin(dlat/2)**2 + np.cos(dec1) * np.cos(dec2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))

    # Calculate the angular distance (in radians)
    distance = c

    # Convert the angular distance from radians to degrees
    distance_deg = np.degrees(distance)

    return distance_deg

# # Example usage:
# ra1 = 50.0  # Replace with the RA of the first point
# dec1 = 35.0  # Replace with the Dec of the first point
# ra2 = 45.0  # Replace with the RA of the second point
# dec2 = 30.0  # Replace with the Dec of the second point
# point1 = [ra1,dec1]
# point2 = [ra2,dec2]
# angular_distance = calculate_distance(point1, point2)
# print(f"Angular distance: {angular_distance} degrees")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_cdf_of_time(points_arrays, likelihoods_arrays, paths_compounded, n_s, cutoff_time, rate, std_dev):
    plt.figure(figsize=(10, 6))
    plt.gca().set_prop_cycle('color', plt.cm.get_cmap('tab10').colors)

    k = 0
    for points_sets, likelihoods_sets in zip(points_arrays, likelihoods_arrays):
        l = 0
        for points, likelihoods in zip(points_sets, likelihoods_sets):
            cdf_values = [0]
            paths = paths_compounded[k][l]
            num_galaxy = len(paths[0])
            for i, path in enumerate(paths):
                times = [0]
                cumulative_prob = likelihoods[path[0]]  # Initialize cumulative probability to 0.0
                current_dist = 0
                current_position = points[path[0]]
                cdf_values = [cumulative_prob]
                for j in range(1, len(path)):
                    next_position = points[path[j]]
                    distance = calculate_distance(np.array(next_position),np.array(current_position))
                    current_dist += distance
                    current_time = convert_distance_to_time(current_dist, rate)
                    times.append(current_time)
                    current_position = next_position
                    cumulative_prob += likelihoods[path[j]]  # Increase cumulative probability by likelihood of the point
                    cdf_values.append(cumulative_prob)

                plt.plot(times, cdf_values, marker='o', linestyle='-', markersize=0.5, linewidth=0.5, label=f'Algorithm {i+1} (n = {n_s[i]})')

                # Find the value of the cumulative probability at cutoff_time
                idx = np.argmin(np.abs(np.array(times) - cutoff_time))
                cdf_cutoff = cdf_values[idx]

                # Display cutoff_time and cdf_cutoff on the graph with adjusted xytext
                offset = 0.08 * i
                plt.scatter(cutoff_time, cdf_cutoff, marker='x', color='red', label='Cutoff Time')
                plt.annotate(f'CDF at Cutoff Time: {cdf_cutoff:.4f}', xy=(cutoff_time, cdf_cutoff), xytext=(cutoff_time + (cutoff_time/10), cdf_cutoff + offset), arrowprops=dict(facecolor='black', arrowstyle='->'))
                                                                                                                               #10
            # Calculate the value of the CDF at the cutoff time
            plt.xlabel('Time')
            plt.ylabel('CDF')
            plt.title(f'Cumulative Distribution Function for set {k+1} and points {l+1}')
            plt.legend(loc = 'upper left', fontsize = 'x-small')

            # Add a box with cutoff_time, standard deviation, and area
            box_text = f'Cutoff Time: {cutoff_time:.1f}\nStandard Deviation: {std_dev:.1f}\nArea: {rate_to_area(rate):.1f}\nNumber of Galaxies: {num_galaxy}'
            plt.text(0.98, 0.02, box_text, transform=plt.gca().transAxes, fontsize=10, verticalalignment='bottom', horizontalalignment='right', bbox=dict(facecolor='white', edgecolor='gray', boxstyle='round,pad=0.5'))

            plt.tight_layout()
            plt.yscale('log')
            plt.xscale('log')
            plt.show()

            l += 1
        k += 1
```
